```python
import allel
import numpy as np
import pandas as pd
```

```python
ace1loc_start = 161486025
ace1loc_end = 161871375
chrom='3'
```

```python
vcf = allel.read_vcf(f"../data/variants/snpEff.merged.vcf.gz", 
                        fields=['variants/*', 'calldata/DP', 
                                'calldata/GT', 'FILTER', 'REF', 'ALT', 'VF'])

chr3 = vcf['variants/CHROM'] == chrom
```

```python
snpEff = pd.DataFrame(vcf['variants/ANN'])[0].str.split("|", expand=True)
snpEff3 = snpEff[chr3]
```

```python
pos = allel.SortedIndex(vcf['variants/POS'][chr3])

ALT = vcf['variants/ALT'][chr3]
vf = vcf['calldata/VF'][chr3][:,0]
dp = vcf['calldata/DP'][chr3]

ace1idx = pos.locate_range(ace1loc_start, ace1loc_end)
pos[ace1idx][snpEff3[ace1idx][1] == 'missense_variant']
```

```python
ace1_mutations = snpEff3[ace1idx][snpEff3[ace1idx][1] == 'missense_variant']

geneid= np.repeat("AAEL000511", ace1_mutations.shape[0])
gene = np.repeat("ACE1", ace1_mutations.shape[0])
ALT = ALT[ace1idx][snpEff3[ace1idx][1] == 'missense_variant'][:,0]
ALT2 = np.repeat("", ace1_mutations.shape[0])
names = ace1_mutations[10].values
locations = [chrom +":"+ p.astype(str) + "-" + p.astype(str) for p in pos[ace1idx][snpEff3[ace1idx][1] == 'missense_variant']]
```

```python
pd.DataFrame([gene, geneid, ALT, ALT2, names, locations]).T.to_csv("Ace1mutations.csv")
```
